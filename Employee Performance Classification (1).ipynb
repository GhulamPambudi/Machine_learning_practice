```python
import pandas as pd
import numpy as np
```

## Load Data

```python
performs = pd.read_excel('dataset_firstmodel_performance.xlsx')
performs
```

```python
performs.head(10)
```

```python
performs2 = performs.sort_values('Type of Performance')
performs2
```

### Converting features and classes into numpy array

```python
features_model = np.array(performs2[['Hard Skill','Soft Skill']])
print('features_model:\n',features_model)
print('')

from sklearn.preprocessing import LabelEncoder

lb = LabelEncoder()
target_model = np.array(performs2['Type of Performance'])
target_model_biner = lb.fit_transform(target_model)
print('target_model_biner:\n',target_model_biner)
```

```python
classes = performs2['Type of Performance'].unique()
print(classes)
```

```python
from sklearn.preprocessing import LabelEncoder

lb = LabelEncoder()
classes_model = lb.fit_transform(classes)
classes_model
```

## Training & Testing Data

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(features_model,target_model_biner,random_state=0)
```

```python
print(X_train)
print(y_train)
```

## Visualizing Data

```python
import matplotlib.pyplot as plt
```

```python
X_train_pandas = pd.DataFrame(X_train)
X_train_pandas['target_inverse'] = lb.inverse_transform(y_train)
X_train_pandas

for i in classes:
    Xpoints = np.array(X_train_pandas[0].loc[X_train_pandas.target_inverse == i])
    ypoints = np.array(X_train_pandas[1].loc[X_train_pandas.target_inverse == i])
    plt.scatter(Xpoints,ypoints,label=i)

plt.legend()
plt.xlabel('Value of Hard Skill')
plt.ylabel('Value of Soft Skill')
plt.title('Performance Classification')
plt.show()
```

## Using K-Nearest Neighbors Model

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train,y_train)
```

## Choosing the Best Value of K

```python
range_n = np.arange(1,11)

ii = []
score_ii_test = []
score_ii_train = []

for i in range_n:
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    knn_test_score = knn.score(X_test,y_test)
    knn_train_score = knn.score(X_train,y_train)
    ii.append(i)
    score_ii_test.append(knn_test_score)
    score_ii_train.append(knn_train_score)

plt.plot(ii,score_ii_test,label='test score')
plt.plot(ii,score_ii_train,label='train_score')

plt.legend()
plt.show()
```

### Since the Best Value of K is about One, Then Remake the K-Nearest Neighbors Model !!
